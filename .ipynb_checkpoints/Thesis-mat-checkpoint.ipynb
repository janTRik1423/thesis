```python
def mat2py(dir, granule):
  if granule[0] == 'T':
    granule = 'A' + granule[1:]
  pp = scipy.io.loadmat(dir+granule+'.full.mat')

  elements = ['B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'B8', 'B9', 'B10', 'B11', 
              'B12', 'B13lo', 'B13hi', 'B14lo', 'B14hi', 'B15', 'B16', 'B17', 
              'B18', 'B19', 'B20', 'B21', 'B22', 'B23', 'B24', 'B25', 'B26', 
              'B27', 'B28', 'B29', 'B30', 'B31', 'B32', 'B33', 'B34', 'B35', 
              'B36', 'GeoFields', 'TrueColor','AOT','BenchMark']
  
  tracks = pp['GeoFields']['Latitude'][0][0].shape[0]
  datadic = {}
  cnt = -1
  data = np.zeros((107,tracks,1354),dtype='float32')

  for band in elements:
    print(band)
    if band == 'GeoFields':
      print('Latitude', pp[band]['Latitude'][0][0].shape)
      print('Longitude', pp[band]['Longitude'][0][0].shape)
      cnt+=1
      data[cnt] = pp[band]['Latitude'][0][0]
      datadic[cnt] = [band, 'Latitude']
      cnt+=1
      data[cnt] = pp[band]['Longitude'][0][0]
      datadic[cnt] = [band, 'Longitude']
    elif band == 'TrueColor':
      print('R', pp[band]['r'][0][0].shape)
      print('G', pp[band]['g'][0][0].shape)
      print('B', pp[band]['b'][0][0].shape)
      print(cnt+1)
      cnt+=1
      data[cnt] = pp[band]['r'][0][0]
      datadic[cnt] = [band, 'r']
      cnt+=1
      data[cnt] = pp[band]['g'][0][0]
      datadic[cnt] = [band, 'g']
      cnt+=1
      data[cnt] = pp[band]['b'][0][0]
      datadic[cnt] = [band, 'b']
    elif band == 'AOT':
      print('Orig', pp[band]['Orig'][0][0].shape)
      print('Corr', pp[band]['Corr'][0][0].shape)
      cnt+=1
      data[cnt] = pp[band]['Orig'][0][0]
      datadic[cnt] = [band, 'Orig']
      cnt+=1
      data[cnt] = pp[band]['Corr'][0][0]
      datadic[cnt] = [band, 'Corr']
    elif band == 'BenchMark':
      print('Ackerman', pp[band]['Ackerman'][0][0].shape)
      print('Hao', pp[band]['Hao'][0][0].shape)
      cnt+=1
      data[cnt] = pp[band]['Ackerman'][0][0]
      datadic[cnt] = [band, 'Ackerman']
      cnt+=1
      data[cnt] = pp[band]['Hao'][0][0]
      datadic[cnt] = [band, 'Hao']
    else:
      print('Rad', pp[band]['Rad'][0][0].shape)
      print('Ref', pp[band]['Ref'][0][0].shape)
      cnt+=1
      data[cnt] = pp[band]['Rad'][0][0]
      datadic[cnt] = [band, 'Rad']
      if np.sum(pp[band]['Ref'][0][0].shape) > 0:
        cnt+=1
        data[cnt] = pp[band]['Ref'][0][0]
        datadic[cnt] = [band, 'Ref']
      print('UI', pp[band]['UI'][0][0].shape)
      cnt+=1
      data[cnt] = pp[band]['UI'][0][0]
      datadic[cnt] = [band, 'UI']


  elements = ['B1hkm','B2hkm']

  datadichkm = {}
  cnt = -1
  datahkm = np.zeros((7,tracks*2,2708),dtype='float32')

  for band in elements:
    print(band)
    if band == 'B1hkm':
      print('TC', pp[band]['TC'][0][0].shape)
      cnt+=1
      datahkm[cnt] = pp[band]['TC'][0][0]
      datadichkm[cnt] = [band, 'TC']    
    print('Rad', pp[band]['Rad'][0][0].shape)
    print('Ref', pp[band]['Ref'][0][0].shape)
    cnt+=1
    datahkm[cnt] = pp[band]['Rad'][0][0]
    datadichkm[cnt] = [band, 'Rad']
    if np.sum(pp[band]['Ref'][0][0].shape) > 0:
      cnt+=1
      datahkm[cnt] = pp[band]['Ref'][0][0]
      datadichkm[cnt] = [band, 'Ref']
    print('UI', pp[band]['UI'][0][0].shape)
    cnt+=1
    datahkm[cnt] = pp[band]['UI'][0][0]
    datadichkm[cnt] = [band, 'UI']


  elements = ['B1qkm','B2qkm']

  datadicqkm = {}
  cnt = -1
  dataqkm = np.zeros((7,tracks*4,5416),dtype='float32')

  for band in elements:
    print(band)
    if band == 'B1qkm':
      print('TC', pp[band]['TC'][0][0].shape)
      cnt+=1
      dataqkm[cnt] = pp[band]['TC'][0][0]
      datadicqkm[cnt] = [band, 'TC']    
    print('Rad', pp[band]['Rad'][0][0].shape)
    print('Ref', pp[band]['Ref'][0][0].shape)
    cnt+=1
    dataqkm[cnt] = pp[band]['Rad'][0][0]
    datadicqkm[cnt] = [band, 'Rad']
    if np.sum(pp[band]['Ref'][0][0].shape) > 0:
      cnt+=1
      dataqkm[cnt] = pp[band]['Ref'][0][0]
      datadicqkm[cnt] = [band, 'Ref']
    print('UI', pp[band]['UI'][0][0].shape)
    cnt+=1
    dataqkm[cnt] = pp[band]['UI'][0][0]
    datadicqkm[cnt] = [band, 'UI']

  np.save(dir+granule+'.full.1km.npy', data)
  np.save(dir+granule+'.full.1km.dictionary.npy', datadic)
  np.save(dir+granule+'.full.hkm.npy', datahkm)
  np.save(dir+granule+'.full.hkm.dictionary.npy', datadichkm)
  np.save(dir+granule+'.full.qkm.npy', dataqkm)
  np.save(dir+granule+'.full.qkm.dictionary.npy', datadicqkm)
```
